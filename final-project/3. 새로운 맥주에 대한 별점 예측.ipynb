## 사용자 평점 예측

개인의 평점을 반영하여 새로운 맥주에 대한 평점 예측

### 코사인 유사도 계산

```python
data = pd.read_csv('data/정제된데이터.csv', encoding='utf-8', index_col=0)
```

```python
# 아이디인 인덱스인 피벗테이블
id_matrix = data.pivot_table('평점', index='아이디', columns='맥주')
id_matrix = id_matrix.fillna(0)
id_matrix.head()
```

```python
# 맥주명이 인덱스인 피벗테이블
beer_matrix = id_matrix.transpose()
beer_matrix.head()
```

```python
# 맥주간 코사인 유사도 계산 후 데이터프레임으로 저장

from sklearn.metrics.pairwise import cosine_similarity

beer_similarity = cosine_similarity(beer_matrix, beer_matrix)
beer_similarity_df = pd.DataFrame(data=beer_similarity, index=id_matrix.columns, columns=id_matrix.columns)

beer_similarity_df.head()
```

```python
print(beer_similarity_df.shape)
```

### 평점 예측 알고리즘

```python
def predict_rating(ratings_arr, item_sim_arr):
    # 실제 매긴 평점과 유사도를 기반으로 평점 예측
    ratings_pred = ratings_arr.dot(item_sim_arr) / np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
```

```python
ratings_pred = predict_rating(id_matrix.values, beer_similarity_df.values)

ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=id_matrix.index, columns = id_matrix.columns)

ratings_pred_matrix.head()
```

#### 아이템 기반 최근접 TOP-N 이웃

```python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
    pred = np.zeros(ratings_arr.shape)

    for col in range(ratings_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
        for row in range(ratings_arr.shape[0]):
            pred[row, col] = item_sim_arr[col,:][top_n_items].dot(
            ratings_arr[row, :][top_n_items].T)
            pred[row, col] /= np.sum(item_sim_arr[col,:][top_n_items])

    return pred

ratings_pred = predict_rating_topsim(id_matrix.values, beer_similarity_df.values, n=10)
print('아이템 기반 최근접 TOP-N 이웃 MSE: ', get_mse(ratings_pred, id_matrix.values))

# 계산된 예측 평점 데이터는 DataFrame으로 재생성
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=id_matrix.index, columns=id_matrix.columns)

ratings_pred_matrix.head()
```

```python
def get_not_tried_beer(id_matrix, user):
    matrix_by_id = id_matrix.loc[user, :] # 해당 사용자의 평점 정보
    not_tried = matrix_by_id[matrix_by_id <= 0].index.tolist() # 마셔보지 않은 맥주명 리스트
    return not_tried
```

```python
def recomm_beer_by_userid(pred_df, user, not_tried, n):
    recomm_beer = pred_df.loc[user, not_tried].sort_values(ascending=False)[:n] # 평점이 높을 상위 n개 출력
    return recomm_beer
```

```python
not_tried = get_not_tried_beer(id_matrix, username)
not_tried
```
